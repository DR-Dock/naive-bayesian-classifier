```python
from __future__ import division
from collections import defaultdict
from math import log
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from collections import Counter
```

```python
def train_new(X):  #функция тренировки модели
    size_data = X.shape[1]#сохраняет размерность
    size_classes = X.shape[2]
    feat = np.sum(X, axis=1)# складывает блок тренировочных данных (сколько раз каждый признак встретился в данном классе)
    teta = (feat + 1) / (size_data + 2)# расчёт вероятностей, прибавлем в числитель 1, в знаменатель 2, что бы не получился 0, вообще в задании про это сказано
    t = size_data / (size_data * size_classes)# опреорная вероятность, кажется так называется
    return teta, t
```

```python
def generation(shp):#генерация меток
    m0 = np.array(([0 for i in range(0, shp)]))
    m1 = np.array(([1 for i in range(0, shp)]))
    m2 = np.array(([2 for i in range(0, shp)]))
    m3 = np.array(([3 for i in range(0, shp)]))
    m4 = np.array(([4 for i in range(0, shp)]))
    m5 = np.array(([5 for i in range(0, shp)]))
    m6 = np.array(([6 for i in range(0, shp)]))
    m7 = np.array(([7 for i in range(0, shp)]))
    m8 = np.array(([8 for i in range(0, shp)]))
    m9 = np.array(([9 for i in range(0, shp)]))
    m=np.concatenate((m0, m1,m2,m3,m4,m5,m6,m7,m8,m9))
    return m
m = generation(600)
m1 =generation(500)
mt= generation(60)
m.shape
```

```python
def NBclassify(classifier, x_test): #функция классификации, принимает классификатор и тестовые данные
    a,b = classifier # берё вероятности
    lg = np.log(a) # сразу же логарифмируем
    lg1 = np.log(1-a) #обратные вероятности для нулевых пикселей
    transp_imgs = x_test.transpose() #транспонирование блока данных, нужно для перемножения
    prg = transp_imgs.dot(lg) #перемножение тестовых с логарифмами
    img = ((transp_imgs - 1)* -1).dot(lg1) #выделяем незасвеченные пиксели и умножаем на оратные вероятности
    v = prg+img+np.log(b) #складываем всё по формуле 
   # return np.argmax(v)+ 1
    m = np.argmax(v, axis=2)#.transpose()
    return m.ravel()
```

```python

def error(thread):
    all_variables_bin =  all_variables['data'] > thread * 255 # бинаризация данных с данным thread
    x_train = all_variables_bin[:,:600,:]# отрезаем данные для тренировки
    x_test = all_variables_bin[:,600:,:]# отрезаем данные для теста
    a,b = train_new(x_train) # тренеруем модель
    pr = NBclassify((a,b), x_test)# классифицируем на тестовых
    pr1 = NBclassify((a,b), x_train)# классифицируем тренировочные
    z = pr - m1#вычитаем из предсказанных исходные, которые мы сгенерировали выше, те что равны 0, это верно спрогнозированные, иначе не верно
    z1 = pr1 - m# так же как и выше
    return (np.count_nonzero(z)/(x_test.shape[1]*10),np.count_nonzero(z1)/(x_train.shape[1]*10),pr1)# расчёт ошибок
```

```python
def chack(all_variables_bin): #перекрёстная проверка
    l=np.array([])
    l1=np.array([])
    for i in thread: #цикл по всем thread
        u=0
        u1=0
        for j in range(0,5):# цикл по всем блокам перекрёстной проверки
            all_variables =  all_variables_bin > i * 255 # бинаризация при i-ом thread
            test = all_variables[:,j::5,:] #разбиение по заданию
            test_y=m[j::5] #берюм так же нужные метки
            #train = all_variables['data'][:,all_variables['data'] != j,:]
            I = np.zeros(all_variables.shape[1]) #манипуляции для получения train данных
            I[j::5]=1 #
            I1 = np.zeros(m.shape[0]) #
            I1[j::5]=1 #
            train = all_variables[:,I!=1,:] #
            train_y = m[I1!=1] #
            a,b = train_new(train) #учим 
            pr = NBclassify((a,b), test) #классифицируем
            z = pr - test_y #берём правильные
            y=1-(np.count_nonzero(z)/(test.shape[1]*10)) #правильность
            u += y#для расчёта среднего
            
            pr1 = NBclassify((a,b), train)#классифицируем train
            z1 = pr1 - train_y#правильность
            y1=1-(np.count_nonzero(z1)/(train.shape[1]*10))#
            u1 += y1#для расчёта среднего
        u /= 5#расчёт среднего по test
        u1 /=5#расчёт среднего по train
        l=np.append(l, u)#добавить в список
        l1=np.append(l1, u1)#
    return l,l1
```

```python
t,t1 = chack(x_train)
print(np.argmax(t))#аргумент у лучшего результата при перекрёстной 
```

```python
q,q1,pr=error(thread[np.argmax(t)]) #считаем ошибки для наилучшего thread который был вычислен при перекрёстной 
print("{0} - точность на тестовом наборе".format(1-q))
print("{0} - точность на тестовом наборе в процентах".format((1-q)*100))
print("{0} - точность на обучающем наборе".format(1-q1))
print("{0} - точность на обучающем наборе в процентах".format((1-q1)*100))
```
